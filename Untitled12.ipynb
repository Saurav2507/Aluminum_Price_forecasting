```python
# %%
# Import necessary libraries
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine, text
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import tensorflow as tf

# Load data
aluminium_data = pd.read_csv("/content/sample_data/final_one_Sql.csv")

# Data preprocessing
aluminium_data['date'] = pd.to_datetime(aluminium_data['date'])
aluminium_data = aluminium_data.sort_values(by='date')
aluminium_data = aluminium_data[['date', 'Price']]

# Removing Outliers using IQR
Q1 = aluminium_data['Price'].quantile(0.25)
Q3 = aluminium_data['Price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
aluminium_data_filtered = aluminium_data[(aluminium_data['Price'] > lower_bound) & (aluminium_data['Price'] < upper_bound)]

# Feature scaling
scaler = MinMaxScaler(feature_range=(0,1))
aluminium_data_filtered['Price_scaled'] = scaler.fit_transform(aluminium_data_filtered[['Price']])
data = aluminium_data_filtered['Price_scaled'].values

# Function to create sequences for LSTM
def create_sequences(data, time_step=30):
    X, Y = [], []
    for i in range(len(data) - time_step):
        X.append(data[i:i + time_step])
        Y.append(data[i + time_step])
    return np.array(X), np.array(Y)

# Create sequences
time_step = 30
X, Y = create_sequences(data, time_step)

# Split data into train and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
Y_train, Y_test = Y[:train_size], Y[train_size:]

# Reshape input for LSTM model
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Build LSTM model
model = Sequential([
    LSTM(100, return_sequences=True, input_shape=(time_step, 1)),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25),
    Dense(1)
])

# Compile model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Train the model
history = model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=100, batch_size=16, verbose=1)

# Make predictions
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Inverse transform predictions to original scale
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)

# Inverse transform Y_test
Y_test = scaler.inverse_transform(Y_test.reshape(-1, 1))

# Plot actual vs predicted prices
plt.figure(figsize=(14,6))
plt.plot(aluminium_data_filtered['date'].iloc[train_size+time_step:], Y_test, label="Actual Price", color='blue')
plt.plot(aluminium_data_filtered['date'].iloc[train_size+time_step:], test_predict, label="Predicted Price", color='red')
plt.xlabel("Date")
plt.ylabel("Aluminum Price")
plt.title("Aluminum Price Forecasting using LSTM")
plt.legend()
plt.show()

# Number of future days to predict
future_days = 60

# Get the last 'time_step' days from the dataset
future_input = data[-time_step:].reshape(1, time_step, 1)

future_predictions = []

for _ in range(future_days):
    predicted_price = model.predict(future_input, verbose=0)
    future_predictions.append(predicted_price[0, 0])
    # We are now using predicted price directly without reshaping it again.
    future_input = np.append(future_input[:, 1:, :], predicted_price.reshape(1, 1, 1), axis=1) # We are now reshaping the predicted price to make it (1,1,1) here rather than before the append.

# Convert predictions back to original scale
future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

# Generate future dates
last_date = aluminium_data_filtered['date'].iloc[-1]
future_dates = [last_date + datetime.timedelta(days=i) for i in range(1, future_days + 1)]

# Plot actual vs predicted prices
plt.figure(figsize=(14,6))

# Plot actual prices
plt.plot(aluminium_data_filtered['date'], aluminium_data_filtered['Price'], label="Actual Price", color='blue')

# Plot predicted prices
plt.plot(future_dates, future_predictions, label="Forecasted Price (2025)", color='green', linestyle='dashed')

# Formatting the plot
plt.xlabel("Date")
plt.ylabel("Aluminum Price")
plt.title("Aluminum Price Forecasting for 2025 using LSTM")
plt.legend()
plt.grid(True)  # Add grid lines
plt.show()
```

```python
# prompt: evaluate matrices on the train data

# Evaluate the model on the training data
train_loss = model.evaluate(X_train, Y_train, verbose=0)
print(f"Train Loss: {train_loss}")
```

```python
# prompt: all evalution matrics on the train data

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calculate evaluation metrics on the training data
train_mse = mean_squared_error(Y_train, train_predict)
train_rmse = np.sqrt(train_mse)
train_mae = mean_absolute_error(Y_train, train_predict)
train_r2 = r2_score(Y_train, train_predict)

print(f"Train MSE: {train_mse}")
print(f"Train RMSE: {train_rmse}")
print(f"Train MAE: {train_mae}")
print(f"Train R-squared: {train_r2}")
```

```python
# Evaluate Model
from sklearn.metrics import mean_absolute_error, mean_squared_error # Import necessary metrics functions

mae = mean_absolute_error(Y_test, test_predict)
mse = mean_squared_error(Y_test, test_predict)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((Y_test - test_predict) / Y_test)) * 100
accuracy = 100 - mape

# Store results in DataFrame
results_df = pd.DataFrame({
    "Metric": ["MAE", "MSE", "RMSE", "MAPE", "Accuracy"],
    "Value": [mae, mse, rmse, mape, accuracy]
})

# Print results
print(results_df)
```
